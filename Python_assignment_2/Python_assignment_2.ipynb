# Reading the UK image as provided

# Adding Limiters

# https://towardsdatascience.com/easy-steps-to-plot-geographic-data-on-a-map-python-11217859a2db  used this article as a refrence

```python
limiter=(-10.592,1.6848, 50.681, 57.985)
```

# Plotting the data onto the map

```python
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(da.Longitude, da.Latitude, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Plotting Spatial Data on Riyadh Map')
ax.set_xlim(limiter[0],limiter[1])
ax.set_ylim(limiter[2],limiter[3])
ax.imshow(a, zorder=0, extent = limiter, aspect= 'equal')
```
